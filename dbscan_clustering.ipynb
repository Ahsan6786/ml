```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("crop_data.csv")

X = df[['Temperature', 'Humidity', 'Soil_Moisture', 'Area', 'Production']]
X = X.dropna()
df = df.loc[X.index]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
db = DBSCAN(eps=0.7, min_samples=5)
df['Cluster'] = db.fit_predict(X_scaled)

print(df.head())
```

```python
df[df['Cluster'] == -1]
```

```python
print(df['Cluster'].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Temperature'], df['Production'], c=df['Cluster'])  #temp x and production y, c is the colour of the cluster according to which the graph is made
plt.show()
```
